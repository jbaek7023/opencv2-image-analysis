```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# read in the image
image = cv2.imread('images/monarch_1.jpg')

# make a copy of the image
image_copy = np.copy(image)

# change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

## Prepare data for k-means

```python
# Reshape image into a 2D array of pixels and 3 color values (RGB)
pixel_vals = image_copy.reshape((-1, 3))
# m = number of pixels, 3 = # of color channels 

# Convert to float type
pixel_vals = np.float32(pixel_vals)
```

## Implment k-means clustering

```python
# Define stopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
# a value of epsilon or a max number of iteration
# max number of iteration is 10
# it stops when it reached over epsilon (convergence)

# Perform k-means clustering
k = 6
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
# m by 3 pixels values
# k value
# None: any labels we want
# criteria : stop criteria
# 10 : the number of attemps and finally how we choose our initial center points.
# how we choose initial center points : randomly

# Convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# Reshape data into the original image dimensiosn
segmented_image = segmented_data.reshape((image_copy.shape))
labels_reshape = labels.reshape(image_copy.shape[0], image_copy.shape[1])

plt.imshow(segmented_image)
```

```python
# visualize one segment
plt.imshow(labels_reshape==1, cmap='gray')
```

```python
# Mask image segment
masked_image = np.copy(image_copy)
masked_image[labels_reshape == 1] = [0, 0, 0]
plt.imshow(masked_image)
```
