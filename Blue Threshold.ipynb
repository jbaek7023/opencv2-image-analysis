# Color Threshold, Blue Screen

## Import resources

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Read in and display the image

```python
# Read in the image
image = cv2.imread('images/p.jpg')

# Print out the type of the image data and its dimension
print('Image type:' ,type(image), ' | SHAPE:',image.shape)
```

```python
%matplotlib inline

#display the image
plt.imshow(image)
```

```python
print(image)
```

```python
#BGR now. We want to cvt it to RGB
image_copy = np.copy(image) #since the image is just np array

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

#Display the image
plt.imshow(image_copy)
```

## Define the color threshold

```python
# Define the threshold boundary
lower_blue = np.array([0, 0, 220]) # define the range... it's up to you! experiment!
upper_blue = np.array([50, 70, 255])
```

## Create a mask

```python
# define the masked area
mask = cv2.inRange(image_copy, lower_blue, upper_blue)

plt.imshow(mask, cmap='gray')

#0 : out of bound
#1 : in the bound
```

```python
# mask the image to let the pizza show through
masked_image = np.copy(image_copy)

masked_image[mask != 0 ] = [0, 0, 0]

plt.imshow(masked_image)
```

## Mask and add background image

```python
background_image = cv2.imread('images/p.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

crop_background = background_image[0:720, 0:1280]

crop_background[mask==0] = [0, 0, 0]

plt.imshow(crop_background)
```

```python
combined_image = crop_background + masked_image
plt.imshow(combined_image)
```
