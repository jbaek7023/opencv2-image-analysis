# Color Spaces and Transform

## Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/water_balloons.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

## Plot color channels

```python
# RGB channels

r = image_copy[:, :, 0]
g = image_copy[:, :, 1]
b = image_copy[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```python
image_copy_hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
plt.imshow(image_copy_hsv)
```

```python
# HSV channels
image_copy_hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)

h = image_copy_hsv[:, :, 0]
s = image_copy_hsv[:, :, 1]
v = image_copy_hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title('Red')
ax1.imshow(h, cmap='gray')

ax2.set_title('Green')
ax2.imshow(s, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(v, cmap='gray')
```

## Define pink and HUE selection thresholds

```python
# Define our color selection criteria in RGB values
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])
```

```python
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

## Mask the image

```python
mask_rgb = cv2.inRange(image_copy, lower_pink, upper_pink)

masked_image = np.copy(image_copy)
masked_image[mask_rgb==0] = [0,0,0]

plt.imshow(masked_image)
plt.title('RGB selection')
```

```python
mask_hue = cv2.inRange(image_copy_hsv, lower_hue, upper_hue)
masked_image2 = np.copy(image_copy)
masked_image2[mask_hue==0] = [0,0,0]

plt.imshow(masked_image2)
plt.title('HUE selection')
```
